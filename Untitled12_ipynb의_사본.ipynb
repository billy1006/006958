```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import ccf
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# 데이터 불러오기
df = pd.read_excel("test_2025-06-22.xlsx")

# 1. 시간 정제 및 변환
df.columns = df.columns.str.strip()
df['시간'] = df['시간'].astype(str).apply(lambda x: re.sub(r'[^\x00-\x7Fㄱ-힣0-9: -]', '', x))
df['시간'] = pd.to_datetime(df['시간'], errors='coerce')
df = df.sort_values('시간').reset_index(drop=True)

# 2. 컬럼 이름 통일
df.rename(columns={
    '강우(mm)': 'rain',
    '집수정유량 (L/min)': 'flow_catchment',
    '원형수로유량 (L/min)': 'flow_pipe'
}, inplace=True)

# 3. 결측치 제거
df = df.dropna(subset=['시간', 'rain', 'flow_catchment', 'flow_pipe'])

# 4. 시각화
plt.figure(figsize=(14, 6))
plt.plot(df['시간'], df['rain'], label='Rain (mm)', alpha=0.5)
plt.plot(df['시간'], df['flow_pipe'], label='Pipe Flow (L/min)', alpha=0.7)
plt.legend()
plt.title("Rainfall vs. Pipe Flow")
plt.xlabel("Time")
plt.ylabel("Values")
plt.show()

# 5. 상관분석
print("Pearson Corr:", df['rain'].corr(df['flow_pipe']))
print("Spearman Corr:", df['rain'].corr(df['flow_pipe'], method='spearman'))

# 6. Cross-Correlation 함수
def plot_cross_correlation(x, y, lags=50):
    x = (x - np.mean(x)) / np.std(x)
    y = (y - np.mean(y)) / np.std(y)
    corr = np.correlate(x, y, mode='full')
    lags_array = np.arange(-len(x) + 1, len(x))
    center = len(corr) // 2
    plt.figure(figsize=(12, 4))
    plt.plot(lags_array[center - lags:center + lags], corr[center - lags:center + lags])
    plt.title('Cross-correlation (Rain → Flow)')
    plt.xlabel('Lag')
    plt.ylabel('Correlation')
    plt.grid()
    plt.show()

plot_cross_correlation(df['rain'].values, df['flow_pipe'].values)

# 7. 선형 회귀 (시간 지연 포함)
df['rain_lag2'] = df['rain'].shift(2)
df_model = df.dropna(subset=['rain_lag2', 'flow_pipe'])
X = df_model[['rain_lag2']]
y = df_model['flow_pipe']
model = LinearRegression().fit(X, y)
y_pred = model.predict(X)
print("R2:", r2_score(y, y_pred))
print("RMSE:", mean_squared_error(y, y_pred, squared=False))

# 8. ARIMAX 모델
model_arimax = SARIMAX(df_model['flow_pipe'], exog=df_model[['rain_lag2']], order=(1, 0, 0))
results = model_arimax.fit(disp=False)
print(results.summary())

# 예측값 시각화
plt.plot(df_model['시간'], df_model['flow_pipe'], label='Actual')
plt.plot(df_model['시간'], results.fittedvalues, label='Predicted')
plt.legend()
plt.title('ARIMAX Prediction')
plt.show()
```
